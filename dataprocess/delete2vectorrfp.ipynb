```python
import pandas as pd
import os
import glob
```

```python
# for two vector

def f(date:str,x):
    rfp = pd.read_csv(date + '/merged/' + str(x) + '/rfptrack.csv')
    print('rfp:',rfp.shape[0])
    aspot = pd.read_csv(date + '/merged/' + str(x) + '/gfptrack.csv')
    index = aspot[aspot['TRACK_ID']=='None'].index.tolist()
    aspot = aspot.drop(index=index)
    del_index = []
    for t in range(0,111):
        ri = rfp[rfp['POSITION_T']==t].index.tolist()
        ai = aspot[aspot['POSITION_T']==t].index.tolist()
        for i in ri:
            for j in ai:
                d = (rfp.loc[i,'POSITION_X'] - aspot.loc[j,'POSITION_X'])**2 + (rfp.loc[i,'POSITION_Y'] - aspot.loc[j,'POSITION_Y'])**2
                if d <= 225  and j not in del_index:
                    del_index.append(j)
    print(len(del_index))
    aspot = aspot.drop(index=del_index)
    aspot.to_csv(date + '/merged/' + str(x) + '/delrfp.csv',index=False)
```
