```python
import tifffile
import numpy as np
import glob
import os
import matplotlib.pyplot as plt
```

```python
date = r'D:/allen/data/mydata/ac_time_seperated/20210608/'
dir_list = glob.glob(date + 'bf/*')
dir_len = len(dir_list) # number of scence we observed
```

```python
for di in range(1,dir_len+1):
    bf = glob.glob(date+'bf/'+str(di)+'/*.tiff')
    cy3 = glob.glob(date+'cy3/'+str(di)+'/*.tiff')
    cy5 = glob.glob(date+'cy5/'+str(di)+'/*.tiff')
    gfp = glob.glob(date+'gfp/'+str(di)+'/*.tiff')
    nuclei = glob.glob(date + 'nuclei/' + str(di) + '/*.tiff')
#     nuclei_b = glob.glob(date+'nuclei_blured/'+str(di)+'/*.tiff') # binary image of nucleus
    pdmito = glob.glob(date+'pdmito/'+str(di)+'/*tiff') # predicted mitochondria
    
    index_list = []
    for i in range(len(bf)):
        s = 't=' + str(i+1)
        for j in range(len(bf)):
            if s in bf[j]: 
                index_list.append(j)
                break
    
    save_dir = date + '/bf/' + str(di) + '/'
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
    for i in range(len(index_list)):
        if i == 0:
            img1 = np.expand_dims(tifffile.imread(bf[index_list[i]]),axis=0)
        else:
            img1 = np.concatenate((img1,np.expand_dims(tifffile.imread(bf[index_list[i]]),axis=0)))
    tifffile.imwrite(save_dir+'merged.tif',img1,photometric='minisblack', resolution=(1./0.260, 1./0.260))
    
    save_dir = date + '/cy3/' + str(di) + '/'
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
    for i in range(len(index_list)):
        if i == 0:
            img2 = np.expand_dims(tifffile.imread(cy3[index_list[i]]),axis=0)
        else:
            img2 = np.concatenate((img2,np.expand_dims(tifffile.imread(cy3[index_list[i]]),axis=0)))
    tifffile.imwrite(save_dir+'merged.tif',img2,photometric='minisblack', resolution=(1./0.260, 1./0.260))
    
    save_dir = date + '/cy5/' + str(di) + '/'
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
    for i in range(len(index_list)):
        if i == 0:
            img3 = np.expand_dims(tifffile.imread(cy5[index_list[i]]),axis=0)
        else:
            img3 = np.concatenate((img3,np.expand_dims(tifffile.imread(cy5[index_list[i]]),axis=0)))
    tifffile.imwrite(save_dir+'merged.tif',img3,photometric='minisblack', resolution=(1./0.260, 1./0.260))
    
    save_dir = date + '/gfp/' + str(di) + '/'
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
    for i in range(len(index_list)):
        if i == 0:
            img4 = np.expand_dims(tifffile.imread(gfp[index_list[i]]),axis=0)
        else:
            img4 = np.concatenate((img4,np.expand_dims(tifffile.imread(gfp[index_list[i]]),axis=0)))
    tifffile.imwrite(save_dir+'merged.tif',img4,photometric='minisblack', resolution=(1./0.260, 1./0.260))
    
    save_dir = date + '/nuclei/' + str(di) + '/'
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
    for i in range(len(index_list)):
        if i == 0:
            img5 = np.expand_dims(tifffile.imread(nuclei[index_list[i]]),axis=0)
        else:
            img5 = np.concatenate((img5,np.expand_dims(tifffile.imread(nuclei[index_list[i]]),axis=0)))
    tifffile.imwrite(save_dir+'merged.tif',img5,photometric='minisblack', resolution=(1./0.260, 1./0.260))
    
#     save_dir = date + 'nuclei_blured/' + str(di) + '/'
#     if not os.path.exists(save_dir):
#         os.makedirs(save_dir)
#     for i in range(len(index_list)):
#         if i == 0:
#             img6 = np.expand_dims(tifffile.imread(nuclei_b[index_list[i]]),axis=0)
#         else:
#             img6 = np.concatenate((img6,np.expand_dims(tifffile.imread(nuclei_b[index_list[i]]),axis=0)))
#     tifffile.imwrite(save_dir+'merged.tif',img6,photometric='minisblack', resolution=(1./0.260, 1./0.260))
    
    save_dir = date + 'pdmito/' + str(di) + '/'
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)
    for i in range(len(index_list)):
        if i == 0:
            img7 = np.expand_dims(tifffile.imread(pdmito[index_list[i]]),axis=0)
        else:
            img7 = np.concatenate((img7,np.expand_dims(tifffile.imread(pdmito[index_list[i]]),axis=0)))
    tifffile.imwrite(save_dir+'merged.tif',img7,photometric='minisblack', resolution=(1./0.260, 1./0.260))
    
    if not os.path.exists(date+'merged/'+str(di)+'/'):
        os.makedirs(date+'merged/'+str(di)+'/')
    bf = np.expand_dims(img1,axis=1)
    cy5 = np.expand_dims(img3,axis=1)
    cy3 = np.expand_dims(img2,axis=1)
    gfp = np.expand_dims(img4,axis=1)
    nu = img5
#     nb = img6 
    pmito = img7
#     img = np.concatenate((nb,nu,cy5,cy3,gfp,bf),axis=1)
    img = np.concatenate((nu,cy5,cy3,gfp,bf,pmito),axis=1)
#     img = np.concatenate((nb,nu,cy5,cy3,gfp,bf,pmito),axis=1) # 全部都有(包含binary nucleus)

    tifffile.imwrite(date+'merged/'+str(di)+'/'+str(di)+'.tif',img,photometric='minisblack', resolution=(1./0.260, 1./0.260))
    print('saved scence {}, shape:{}'.format(di,img.shape))
    
# shape:(timepoints, channels, y resolution, x resolution)
```
