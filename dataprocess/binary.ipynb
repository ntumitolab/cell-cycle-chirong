```python
from aicsimageio import AICSImage
import tifffile as tf
import numpy as np
import os 
import dask.array as da
import glob
from scipy.ndimage import gaussian_filter as gf
from scipy.ndimage import median_filter as mf
import matplotlib.pyplot as plt
```

```python
d = r"D:/allen/data/mydata/ac_time_seperated/20220412/nuclei/*/*.tiff"
```

```python
file = glob.glob(d)
```

```python
file
```

```python
save_dir = r'D:/allen/data/mydata/ac_time_seperated/20220412/nuclei_blured/'
if not os.path.exists(save_dir):
    os.makedirs(save_dir)
```

```python
for i in range(len(file)):
    img = tf.imread(file[i])
    name = file[i].split('/')[-1].replace('nuclei','')
    if not os.path.exists(save_dir+name.split('\\')[1]):
        os.makedirs(save_dir+name.split('\\')[1])
    imgb = (img > (img.mean()+0.5*img.std())).astype(np.uint8)*255
    tf.imwrite(save_dir+name,imgb,photometric='minisblack')
```
